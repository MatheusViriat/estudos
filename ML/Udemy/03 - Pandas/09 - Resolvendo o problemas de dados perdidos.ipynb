```python
import pandas as pd
import numpy as np
```

```python
dados = {
    'nome' : ['André', 'Julia', 'Ismael', np.nan, 'Jéssica', 'Evelyn', 'Pedrinho'],
    'sexo' : ['M', 'F', 'M', np.nan, 'F', 'F', np.nan],
    'idade' : [28, 53, np.nan, np.nan, 26, 28, 1],
    'nota' : [4, 10, 8, np.nan, 10, 7, 3]
}
df = pd.DataFrame(dados)
```

```python
df
```

```python
#remove os exemplos que tem um ou mais atributos incompletos
df.dropna()
```

```python
#remove os exemplos que tem todos os atributos incompletos
df.dropna(how='all')
```

```python
df['serie'] = np.nan
```

```python
df
```

```python
#remove a coluna onde se encontra todos os atributos incompletos
df.dropna(how='all', axis=1)
```

```python
#remove exemplo que tem 3 ou mais atributos incompletos
df.dropna(thresh=3)
```

```python
#preenchendo NaN da coluna série com 8
df['serie'].fillna(8)
```

```python
#inserindo de fato no df com inplace
df['serie'].fillna(8, inplace=True)
```

```python
df
```

```python
df['serie'] = np.nan
```

```python
df
```

```python
#preenchendo idades NaN com a média de idade
df['idade'].fillna(df['idade'].mean(), inplace=True)
```

```python
#filtrando dados onde o nome nao seja null e nem o sexo
df[df['nome'].notnull() & df['sexo'].notnull()]
```
