```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
salarios_matheus = np.array([100,200,400,300])
salarios_andre = np.array([50, 300, 500, 450])
```

```python
#adicionando label
plt.plot(salarios_matheus, c='Red', ls='--', marker='s', ms=8, label='Salários Matheus')
plt.plot(salarios_andre, c='Black', ls='--', marker='s', ms=8, label='Salários André')
plt.show()
```

```python
#colocando legenda
plt.plot(salarios_matheus, c='Red', ls='--', marker='s', ms=8, label='Salários Matheus')
plt.plot(salarios_andre, c='Black', ls='--', marker='^', ms=8, label='Salários André')
plt.legend()
plt.show()
```

```python
#colocando legenda no canto superior direito
plt.plot(salarios_matheus, c='Red', ls='--', marker='s', ms=8, label='Salários Matheus')
plt.plot(salarios_andre, c='Black', ls='--', marker='^', ms=8, label='Salários André')
plt.legend(loc='upper right')
plt.show()
```

```python
#colocando mais uma linha
salarios_pedro = [200, 150, 300, 700]
plt.plot(salarios_matheus, c='Red', ls='--', marker='s', ms=10, label='Salários Matheus')
plt.plot(salarios_andre, c='Black', ls='--', marker='^', ms=10, label='Salários André')
plt.plot(salarios_pedro, c='Green', ls='--', marker='o', ms=10, label='Salários Pedro')
plt.legend(loc='upper left')
plt.show()
```
