```python
import numpy as np
```

```python
#carregando base de dados
data = np.genfromtxt('iris.data', delimiter=',', usecols=(0,1,2,3))
data
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#comprimeiro da sépala nos exemplos iris-setosa
plt.plot(data[:50,0], c='Red', ls=':', marker='s', ms=8)
```

```python
data[50:100,0]
```

```python
#comprimento das sépalas dos exemplos iris-setosa e iris-versicolor
plt.plot(data[:50,0], c='Red', ls=':', marker='s', ms=8, label='Comp. Sépala Iris-Setosa')
plt.plot(data[50:100,0], c='Black', ls=':', marker='o', ms=8, label='Comp. Sépala Iris-Versicolor')
plt.legend()
plt.show()
```

```python
#comprimento das sépalas dos exemplos iris-setosa, iris-versicolor e iris virginica
plt.plot(data[:50,0], c='Red', ls=':', marker='s', ms=6, label='Comp. Sépala Iris-Setosa')
plt.plot(data[50:100,0], c='Black', ls=':', marker='o', ms=6, label='Comp. Sépala Iris-Versicolor')
plt.plot(data[100:150,0], c='Green', ls=':', marker='^', ms=6, label='Comp. Sépala Iris-Virginica')
plt.legend()
plt.show()
```

```python
#atributos dos exemplos iris-setosa
plt.plot(data[:50,0], c='Red', ls=':', marker='s', ms=6, label='Comp. Sépala')
plt.plot(data[:50,1], c='Black', ls=':', marker='o', ms=6, label='Tamanho Sépala')
plt.legend()
plt.show()
```
