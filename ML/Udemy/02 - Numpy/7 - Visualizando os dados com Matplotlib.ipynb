```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
salario = np.array([100, 200, 300, 400, 500, 450, 150])
salario
```

```python
plt.plot(salario)
```

```python
plt.plot(salario, c='Black')
```

```python
plt.plot(salario, c='Red', ls='--', marker='s')
```
